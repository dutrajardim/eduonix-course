```python
import pandas as pd
import numpy as np
import config

key = config.google_spread_sheets['cemig']['key']
sheet_name = config.google_spread_sheets['cemig']['sheets']
link = f'https://docs.google.com/spreadsheets/d/{key}/gviz/tq?tqx=out:csv&sheet={sheet_name}'

data = pd.read_csv(link, dtype={ 'Instalação': np.str, 'Contrato': np.str }, parse_dates=['Vencimento', 'Mês/Ano'])
data.head()
```

```python
data['Contrato'].fillna(data['Instalação'], inplace=True)
data['Valor'] = data['Valor'].apply(lambda x: x.replace('R$ ', '').replace('.', '').replace(',','.')).astype('float')
```

```python
table = pd.pivot_table(data, values='Valor', index=['Mês/Ano', 'Contrato'], aggfunc=np.sum)
table.loc['2019-10']
```
